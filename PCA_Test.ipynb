```python
import numpy as np
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition, datasets
```

```python
np.random.seed(10)
```

```python
iris_dataset = datasets.load_iris()
X = iris_dataset.data
y = iris_dataset.target
```

```python
figure = pyplot.figure(1, figsize=(4,3))
pyplot.clf()
axes = Axes3D(figure, rect=[0,0,0.95,1],elev= 48, azim=134)
pyplot.cla()

pca = decomposition.PCA(n_components=3)
pca.fit(X)
X = pca.transform(X)

for name, label in [('Setosa', 0), ('Versicolor', 1), ('Virginica', 2)]:
    axes.text3D(X[y == label, 0].mean(),
               X[y == label, 1].mean() + 1.5,
               X[y == label, 2].mean(),
               name,
               horizontalalignment='center',
               bbox=dict(alpha=0.5, edgecolor='w', facecolor='w'))
    
y = np.choose(y, [1, 2, 0]).astype(np.float)

axes.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=pyplot.cm.nipy_spectral, edgecolor='k')
axes.w_xaxis.set_ticklabels([])
axes.w_yaxis.set_ticklabels([])
axes.w_zaxis.set_ticklabels([])
pyplot.show()
```
