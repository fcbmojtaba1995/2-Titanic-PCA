```python
import pandas as pd
import numpy as np
from sklearn import decomposition
from matplotlib import pyplot as plt

data = pd.read_csv('titanic.csv')
data.Age = data.Age.fillna(-0.5)
data = data.drop(['Sex','Fare', 'Pclass','Cabin','Ticket', 'Name' , 'Embarked' ], axis=1)
X = data.drop(['Survived','PassengerId'],axis=1)
y = data['Survived']

pca = decomposition.PCA(n_components=2)
pca.fit(X)
np.random.seed(10)
X = pca.transform(X)

fig, ax = plt.subplots()
y = np.choose(y, [1,0]).astype(np.float)
ax.scatter(X[:,0], X[:,1], c= y , cmap=plt.cm.flag, edgecolor='b')
```
